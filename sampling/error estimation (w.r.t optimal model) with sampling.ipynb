```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pylab
from scipy import stats
from IPython import display
import time
from matplotlib.patches import Rectangle
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python

m = 500
#n = 50
I = 10
r = 10
weighted = False
noise = 50

f = lambda x : -10*x + 3 + 10*np.random.normal(0,noise,1)
x = np.random.normal(0, 10, m)
y = np.array([f(i) for i in x])
x = x.reshape((m,))
y = y.reshape((m,))
bias = np.ones(m)
X = np.vstack((bias,x)).T

lm = LinearRegression()
lm.fit(X,y)

y_pred = lm.predict(X)
e = (y - y_pred)**2
w = e/np.sum(e)
err_base = np.mean(e)

a = lm.coef_
line = a[1]*x + a[0] 
plt.scatter(x,y,alpha=0.5)
plt.plot(x, line, color="orange")
population_err = np.linalg.norm(y - np.dot(X,a))/len(X)
```

```python
# uniform sampling
unif_errors = []
for i in range(I):
    plt.figure(figsize=(10,10))
    plt.plot(x,line,color='orange')
    plt.scatter(x,y,alpha=0.1, s = 100, color = "blue")
    idx = np.random.randint(0,len(X),size=r)
    X1 = X[idx,:]
    y1 = y[idx]
    err = np.mean((lm.predict(X1) - y1)**2)
    unif_errors.append(err)
    plt.scatter(x[idx],y[idx],s=80, color = "red")
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(1)
    plt.clf()

    
#print imp_errors, unif_errors
#print "base error: %.2f" % err_base
#print "variance importance %.2f" % np.var(imp_errors)
#print "variance uniform %.2f" % np.var(unif_errors)
```

```python
# importance sampling
imp_errors = []
for i in range(I):
    plt.figure(figsize=(10,10))
    plt.plot(x,line,color='orange')
    plt.scatter(x,y,alpha=0.1, s = 100, color = "blue")
    idx = np.random.randint(0,len(X),size=r)
    w1 = w[idx]
    if weighted:
        X1 = X[idx,:]/(m*w1[:,None])
        y1 = y[idx]/(m*w1[:,None])
    else:
        X1 = X[idx,:]
        y1 = y[idx]
    e = np.mean((lm.predict(X1) - y1)**2)
    imp_errors.append(e)
    #extra = Rectangle((0, 0), 1, 1, fc="w", fill=False, edgecolor='none', linewidth=0)
    extra = Rectangle((0, 0), 1, 1, fc="w")
    plt.legend([extra],['sample error: '+str(e)])
    plt.scatter(x[idx],y[idx],s=80, color = "red")
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(1)
    plt.clf()
```
